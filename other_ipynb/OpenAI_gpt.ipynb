#Is posts disscuss about an issue or not?

```python
import openai
import pandas as pd

# Set API Key (recommended: use environment variables instead)
openai.api_key = "Insert OpenAI Key"

def is_post_discussing_an_issue(text):
    """Check if a Stack Overflow post is discussing a problem (returns 'yes' or 'no')."""

    messages = [
        {"role": "system", "content": "You are a helpful assistant that checks whether a Stack Overflow post is discussing a problem or not."},
        {"role": "user", "content": f"Is the following post discussing a problem?\n\n{text}\n\nRespond only with 'yes' or 'no'."}
    ]

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=messages,
            max_tokens=5,  # Limit token usage to reduce costs
            temperature=0  # Set to 0 for consistent results
        )

        result = response['choices'][0]['message']['content'].strip().lower()

        return 'yes' if result == 'yes' else 'no'  # Ensure consistent output

    except Exception as e:
        print(f"Error: {e}")
        return 'no'  # Default to 'no' on errors

# Apply function to 'body' column
df['discussing_issue'] = df['body'].apply(is_post_discussing_an_issue)

# Display results
print(df[['title', 'discussing_issue']])
```

```python
df.to_csv('output_issue_or_not.csv', index=False)
```

```python
import pandas as pd

# Load CSV file into a DataFrame
csv_file_path = "/content/bquxjob_7eb45a99_19482c81e45.csv"  # Replace with your CSV file path
df = pd.read_csv(csv_file_path)

# Preview the data
print(df.head())
```

```python
# Extract the column with text data
texts = df['tag_name'].tolist()  # Replace 'text' with the actual column name in your CSV
```

```python
df['text'] = df['tag_name']
```

```python
import openai

# Set your OpenAI API key
openai.api_key = "Insert OpenAI Key"

# Function to check if text is about green software engineering
def classify_green_topic(text):
    prompt = (
        f"The following text discusses a topic. Decide if it is about green software engineering. "
        f"Answer with 'Yes' if it is related to green software engineering and 'No' otherwise. "
        f"Text: {text}\n\nIs this about green software engineering?"
    )
    response = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}]
    )
    result = response['choices'][0]['message']['content'].strip()
    return result

# Apply the classification to all rows
df['is_green_software_engineering'] = df['tag_name'].apply(classify_green_topic)

# Save the results back to a CSV file
df.to_csv("classified_topics.csv", index=False)
print("Classification complete. Results saved to classified_topics.csv.")
```
